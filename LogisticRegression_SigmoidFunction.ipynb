```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# NumPy has a function called exp(), which offers a convenient way to calculate the exponential (ùëí^ùëß
# of all elements in the input array (z).

# Input is an array
input_array = np.array([1,2,3])
exp_array = np.exp(input_array)

print("input to exp", input_array)
print("output of exp", exp_array)

# input is a single number
input_val = 1 
exp_val = np.exp(input_val)

print("input to exp", input_val)
print("output of exp", exp_val)
```

```python
def sigmoid(z):
    """
    Compute the sigmoid of z

    Args:
        z (ndarray): A scalar, numpy array of any size.

    Returns:
        g (ndarray): sigmoid(z), with the same shape as z
         
    """
    
    g = 1/(1+np.exp(-z))
    
    return g
```

```python
# Generate an array evenly spaced values between -10 and 10
z_tmp = np.arange(-10,11)

# Us the function implemented above to calculate the sigmoid of the array
y = sigmoid(z_tmp)

# Code for printting the two arrays next to each other
np.set_printoptions(precision=3)
print("Input(z), Output (sigmoid(z))")
print(np.c_[z_tmp, y])
```

```python
# Plot z vs the sigmoid(z)

fig, ax = plt.subplots()
ax.plot(z_tmp, y, c = 'b')

ax.set_title("Sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
```

# Decision Boundary

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1) # reshapes to shape (6,1)
```

```python
# plot the logistic regression data with The data points with label  ùë¶=1
#  are shown as red crosses, while the data points with label  ùë¶=0
#  are shown as blue circles

fig, ax = plt.subplots(1,1)
ax.scatter(X[y[:,0]==1,0], X[y[:,0]==1,1], c='r', marker='x', label='y=1')
ax.scatter(X[y[:,0]==0,0], X[y[:,0]==0,1], c='b', marker='o', label='y=0')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.axis([0, 4, 0, 3.5])
ax.legend()
```

```python
# Plot sigmoid(z) over a range of values from -10 to 10

z = np.arange(-10,11)
fig, ax = plt.subplots(1,1,figsize=(5,3))
ax.plot(z, sigmoid(z), c='b')

ax.set_title("Sigmoid function")
ax.set_ylabel('sigmod(z)')
ax.set_xlabel('z')
```

```python
# Choose values between 0 and 6
x0 = np.arange(0,6)

x1 = 3 - x0
fig, ax = plt.subplots(1,1,figsize=(5,4))
# Plot the decison boundary
ax.plot(x0, x1, c='b')
ax.axis([0, 6, 0, 6])

# Fill the region below the line
ax.fill_between(x0,x1, alpha=0.2)

# Plot the original data
ax.scatter(X[y[:,0]==1,0], X[y[:,0]==1,1], c='r', marker='x', label='y=1')
ax.scatter(X[y[:,0]==0,0], X[y[:,0]==0,1], c='b', marker='o', label='y=0')
ax.set_ylabel(r'$x_1$')
ax.set_xlabel(r'$x_0$')
plt.show()
```
