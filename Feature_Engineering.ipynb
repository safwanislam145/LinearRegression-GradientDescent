```python
# Feature engineering for the housing data set to predict house prices

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Polinamoial Regression 

```python
# create target dataset

x = np.arange(0, 20, 1)
y = 1 + x**2
X = x.reshape(-1,1) # reshape to 2D array
print(f"X shape: {X.shape}")
print(X)

```

```python
def run_gradient_descent_feng(X, y, iterations=1000, alpha=1e-2):
    m, n = X.shape
    w = np.zeros((n, 1))
    b = 0
    y = y.reshape(-1, 1)
    
    for i in range(iterations):
        # Compute predictions
        y_pred = X @ w + b
        
        # Compute the error
        error = y_pred - y
        
        # Compute the gradients
        dw = (1/m) * (X.T @ error)
        db = (1/m) * np.sum(error)
        
        # Update the weights and bias
        w -= alpha * dw
        b -= alpha * db
        
    return w, b
```

```python
model_w, model_b = run_gradient_descent_feng(X, y)

# Plot the result
plt.scatter(X, y, label='Actual value')
plt.title('no feature engineering')

plt.plot(X, np.dot(X, model_w) + model_b, color='red', label='Predicted value')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Feature engineering
X = x**2
y = 1 + x**2
X = X.reshape(-1,1) # reshape to 2D array

model_w, model_b = run_gradient_descent_feng(X, y)
plt.scatter(x,y, label='Actual value')
plt.title('Added x**2 feature')

plt.plot(X, np.dot(X, model_w) + model_b, color='red', label='Predicted value')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
